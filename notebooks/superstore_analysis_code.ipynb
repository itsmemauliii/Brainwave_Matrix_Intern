## Dataset Description → Superstore Sales Data

The dataset contains transactional records of a commercial store, including customer orders, shipping details, product information, and financial figures. It is commonly used in data analysis and visualization projects to understand retail performance, customer behavior, and logistics.

**Key Columns in the Dataset:**
- **Order ID**: Unique identifier for each order.
- **Order Date / Ship Date**: Dates of order and shipment.
- **Ship Mode**: Shipping method selected.
- **Customer ID / Name**: Identifiers and names of customers.
- **Segment**: Customer segment (Consumer, Corporate, Home Office).
- **City / State / Region / Country**: Geographical details.
- **Product ID / Category / Sub-Category**: Product classification.
- **Sales**: Revenue generated from each order.
- **Quantity**: Number of units sold.
- **Discount**: Discount applied.
- **Profit**: Net profit earned from each order.

This dataset is instrumental in uncovering sales trends, understanding customer segmentation, and identifying profit-generating regions and products.

## Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Load the dataset

```python
data = pd.read_csv("superstore.csv", encoding='ISO-8859-1')
data
```

## Data Overview and Cleaning

```python
print("Missing Values:\n", data.isnull().sum())
print("Duplicates:", data.duplicated().sum())
```

```python
data.drop_duplicates(inplace=True)
```

```python
data['Order Date'] = pd.to_datetime(data['Order Date'])
data['Ship Date'] = pd.to_datetime(data['Ship Date'])
```

```python
data['Month_Year'] = data['Order Date'].dt.to_period('M')
```

## Total Revenue, Profit, and Order Summary

```python
total_sales = data['Sales'].sum()
total_profit = data['Profit'].sum()
total_orders = data['Order ID'].nunique()
print(f"Total Sales: ${total_sales:,.2f}")
print(f"Total Profit: ${total_profit:,.2f}")
print(f"Total Orders: {total_orders}")
```

## Sales by Category

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Category', y='Sales', data=data, estimator=sum)
plt.title("Sales by Category")
plt.show()
```

## Profit by Region

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Region', y='Profit', data=data, estimator=sum)
plt.title("Profit by Region")
plt.show()
```

## Top 10 products by sales

```python
top_products = data.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products.plot(kind='barh', title="Top 10 Products by Sales", figsize=(8,5), color='red')
plt.gca().invert_yaxis()
plt.show()
```

## Monthly Sales Trend

```python
monthly_sales = data.groupby('Month_Year')['Sales'].sum().reset_index()
monthly_sales['Month_Year'] = monthly_sales['Month_Year'].astype(str)

plt.figure(figsize=(10,5))
sns.lineplot(x='Month_Year', y='Sales', data=monthly_sales, marker='o')
plt.xticks(rotation=45)
plt.title("Monthly Sales Trend")
plt.tight_layout()
plt.show()
```

## Discount vs Profit

```python
plt.figure(figsize=(6,4))
sns.scatterplot(data=data, x='Discount', y='Profit', hue='Category')
plt.title("Discount vs Profit")
plt.show()
```

## Sales & Profit by Sub-Category

```python
subcat_data = data.groupby('Sub-Category')[['Sales', 'Profit']].sum().sort_values(by='Sales', ascending=False)

subcat_data.plot(kind='bar', figsize=(10,5), title='Sales & Profit by Sub-Category')
plt.ylabel("Amount ($)")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Profit by Region and Category

```python
pivot = data.pivot_table(values='Profit', index='Region', columns='Category', aggfunc='sum')

plt.figure(figsize=(8,5))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Profit by Region and Category")
plt.show()
```

## Ship Mode Distribution

```python
data['Ship Mode'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(5,5), title='Ship Mode Distribution')
plt.ylabel('')
plt.show()
```

## Profit Distribution by Category

```python
plt.figure(figsize=(6,4))
sns.boxplot(x='Category', y='Profit', data=data)
plt.title("Profit Distribution by Category")
plt.show()
```

## Top 10 Cities by Sales

```python
top_cities = data.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)

top_cities.plot(kind='bar', color='orange', figsize=(8,5), title='Top 10 Cities by Sales')
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.show()
```

## Sales by Region & State Treemap

```python
import plotly.express as px

fig = px.treemap(data, path=['Region', 'State'], values='Sales', title="Sales by Region & State Treemap")
fig.show()
```

## Profit vs Sales (Bubble size = Quantity)

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=data, x='Sales', y='Profit', size='Quantity', hue='Category', alpha=0.6)
plt.title("Profit vs Sales (Bubble size = Quantity)")
plt.show()
```

## Sales by State (USA)

```python
import plotly.express as px

fig = px.choropleth(data,
                    locations="State",
                    locationmode="USA-states",
                    color="Sales",
                    scope="usa",
                    color_continuous_scale="Viridis",
                    title=" Sales by State (USA)")
fig.show()
```

## Sales Distribution: Region > Category > Sub-Category

```python
fig = px.sunburst(data,
                  path=['Region', 'Category', 'Sub-Category'],
                  values='Sales',
                  title='Sales Distribution: Region > Category > Sub-Category')
fig.show()
```

## Sales and Profit Over Time (Monthly)

```python
data_monthly = data.groupby(data['Order Date'].dt.to_period('M')).agg({'Sales': 'sum', 'Profit': 'sum'}).reset_index()
data_monthly['Order Date'] = data_monthly['Order Date'].astype(str)

fig = px.line(data_monthly, x='Order Date', y=['Sales', 'Profit'],
              title="Sales and Profit Over Time (Monthly)",
              labels={'value': 'Amount ($)', 'Order Date': 'Month'})
fig.show()
```

## Sales by Category (Donut Chart)

```python
cat_sales = data.groupby('Category')['Sales'].sum()

fig = px.pie(names=cat_sales.index, values=cat_sales.values, hole=0.5,
             title="Sales by Category (Donut Chart)")
fig.update_traces(textposition='inside', textinfo='percent+label')
fig.show()
```

## Animated Sales by Sub-Category Over Time

```python
data['Month'] = data['Order Date'].dt.to_period('M').astype(str)
animated_data = data.groupby(['Month', 'Sub-Category'])['Sales'].sum().reset_index()

fig = px.bar(animated_data,
             x='Sub-Category',
             y='Sales',
             color='Sub-Category',
             animation_frame='Month',
             range_y=[0, animated_data['Sales'].max() + 500],
             title="Animated Sales by Sub-Category Over Time")
fig.show()
```

## Sales Trend per Sub-Category by Category

```python
fig = px.line(data, x="Order Date", y="Sales", color="Sub-Category",
              facet_col="Category", facet_col_wrap=2,
              title="Sales Trend per Sub-Category by Category")
fig.show()
```

## Heatmap of Sales by Day and Month

```python
data['Day'] = data['Order Date'].dt.day_name()
data['Month'] = data['Order Date'].dt.month_name()

pivot = data.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='sum')

plt.figure(figsize=(12,6))
sns.heatmap(pivot, cmap="coolwarm", annot=True, fmt='.0f')
plt.title("Heatmap of Sales by Day and Month")
plt.show()
```
